```python
import pandas as pd
import requests
from pprint import pprint
import warnings
warnings.filterwarnings('ignore')
```

```python
url = "https://api.themoviedb.org/3/search/movie"
from config import api_key
```

```python
imdb_file = "../Ivan/IMDb_Data_final.csv"
imdb_df = pd.read_csv(imdb_file)
imdb_df.head()
```

```python
imdb_df.info()
```

```python
imdb_film_list = imdb_df["Title"]
imdb_dates = imdb_df["ReleaseYear"]
imbd_film_combo= {}
for i in range(0,len(imdb_film_list)):
    imbd_film_combo[imdb_film_list[i]] = imdb_dates[i]
```

```python
imdb_tmdb_df=pd.DataFrame()

for film in imdb_film_list:
    query_url = f"{url}?api_key={api_key}&query={film}"
    response = requests.get(query_url).json()
    imdb_tmdb_df = imdb_tmdb_df.append(response["results"], ignore_index=True)

imdb_tmdb_df.head()
```

```python
imdb_tmdb_df.info()
```

```python
imdb_tmdb_df["Year"] = imdb_tmdb_df["release_date"].str[:4]
```

```python
filtered_imdb_df = pd.DataFrame()

for title in imbd_film_combo.keys():
    entry = imdb_tmdb_df[(imdb_tmdb_df["original_title"] == title) & (imdb_tmdb_df["Year"] == str(imbd_film_combo[title]))]
    frames = [filtered_imdb_df,entry]
    filtered_imdb_df = pd.concat(frames)

filtered_imdb_df.info()
```

```python
filtered_imdb_df.head()
```

```python
filtered_imdb_df.reset_index(drop=True,inplace=True)
```

```python
filtered_imdb_df.head()
```

```python
filtered_imdb_df['title'].value_counts(ascending=False)
```

```python
drop_dup_df=filtered_imdb_df.drop_duplicates(subset='title')
```

```python
drop_dup_df.info()
```

```python
drop_dup_df.reset_index(drop=True, inplace=True)
```

```python
id_list = drop_dup_df['id']
```

```python
id_list[0]
```

```python
details_url = "https://api.themoviedb.org/3/movie/"
```

```python
details_df= pd.DataFrame()

for id in id_list:
    query_url = f"{details_url}{id}?api_key={api_key}"
    response = requests.get(query_url).json()
    details_df = details_df.append(response, ignore_index=True)

details_df.head()
```

```python
details_df.info()
```

```python
final_df=details_df[['popularity','budget','revenue','genres','original_title','release_date','runtime','production_countries','tagline']]
```

```python
final_df.head(1)
```

```python
final_df['genre']=final_df['genres'][0][0]['name']
```

```python
final_df['movies_country']=final_df['production_countries'][0][0]['name']
```

```python
final_df.columns
```

```python
final_clean_df = final_df[['popularity', 'budget', 'revenue', 'original_title',
       'release_date', 'runtime', 'tagline', 'genre',
       'movies_country']]
```

```python
final_clean_df.rename(columns={'original_title':'title','runtime':'run time','release_date':'release date'}, inplace=True)
```

```python
final_clean_df['release date'] = pd.to_datetime(final_clean_df['release date'])
```

```python
final_clean_df.to_csv("Top_748_Cleanset.csv")
```
