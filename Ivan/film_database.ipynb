```python
import pandas as pd
import requests
from pprint import pprint
import warnings
warnings.filterwarnings('ignore')
```

```python
film_file = "../Ivan/Top 100 Movies of the Last 20 Years.csv"
film_df = pd.read_csv(film_file)
```

```python
film_df.head()
```

```python
film_list = film_df["Title"]
print(film_list[0])
```

```python
film_dates = film_df["Year"]
film_combo = {}
for i in range (0,100):
    film_combo[film_list[i]] = film_dates[i]
```

```python
url = "https://api.themoviedb.org/3/search/movie"
from config import api_key
tmdb_df = pd.DataFrame()


query_url = f"{url}?api_key={api_key}&query=The Shawshank Redemption"
articles = requests.get(query_url).json()
pprint(articles)
```

```python
tmdb_df = tmdb_df.append(articles["results"], ignore_index=True)
tmdb_df.head()
```

```python
tmdb_df = pd.DataFrame()

for film in film_list:
    query_url = f"{url}?api_key={api_key}&query={film}"
    response = requests.get(query_url).json()
    tmdb_df = tmdb_df.append(response["results"], ignore_index=True)

tmdb_df.head()
```

```python
tmdb_df.info()
```

```python
tmdb_df.tail()
```

```python
tmdb_df["Year"] = tmdb_df["release_date"].str[:4]
tmdb_df.head()
```

```python
type(str(film_combo["The Shawshank Redemption"]))
```

```python
filtered_film_df = pd.DataFrame()

for title in film_combo.keys():
    entry = tmdb_df[(tmdb_df["original_title"] == title) & (tmdb_df["Year"] == str(film_combo[title]))]
    frames = [filtered_film_df,entry]
    filtered_film_df = pd.concat(frames)

filtered_film_df.info()
```

```python
imdb_file = "../Ivan/IMDb_Data_final.csv"
imdb_df = pd.read_csv(imdb_file)
imdb_df.head()
```

```python
imdb_df.info()
```

```python
imdb_df.fillna("N/A",inplace=True)
```

```python
imdb_film_list = imdb_df["Title"]
imdb_dates = imdb_df["ReleaseYear"]
imbd_film_combo= {}
for i in range(0,len(imdb_film_list)):
    imbd_film_combo[imdb_film_list[i]] = imdb_dates[i]

```

```python
imdb_tmdb_df=pd.DataFrame()

for film in imdb_film_list:
    query_url = f"{url}?api_key={api_key}&query={film}"
    response = requests.get(query_url).json()
    imdb_tmdb_df = imdb_tmdb_df.append(response["results"], ignore_index=True)

imdb_tmdb_df.head()
```

```python
imdb_tmdb_df.info()
```

```python
imdb_tmdb_df["Year"] = imdb_tmdb_df["release_date"].str[:4]
```

```python
filtered_imdb_df = pd.DataFrame()

for title in imbd_film_combo.keys():
    entry = imdb_tmdb_df[(imdb_tmdb_df["original_title"] == title) & (imdb_tmdb_df["Year"] == str(imbd_film_combo[title]))]
    frames = [filtered_imdb_df,entry]
    filtered_imdb_df = pd.concat(frames)

filtered_imdb_df.info()
```
